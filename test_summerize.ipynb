```python
import gradio as gr
import requests
from bs4 import BeautifulSoup
import urllib.parse
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')

def analyze_sentiment(text, url):
    if url:
        try:
            page = requests.get(url)
            soup = BeautifulSoup(page.content, 'html.parser')
            text = soup.get_text()
        except:
            return "Invalid URL"

    analyzer = SentimentIntensityAnalyzer()
    sentiment = analyzer.polarity_scores(text)

    positive_score = sentiment['pos']
    negative_score = sentiment['neg']
    neutral_score = sentiment['neu']
    compound_score = sentiment['compound']
    polarity_score = 'positive' if compound_score > 0 else 'negative' if compound_score < 0 else 'neutral'

    return pd.DataFrame({
        "Positive Score": [positive_score],
        "Negative Score": [negative_score],
        "Polarity Score": [polarity_score],
        "Neutral Score": [neutral_score],
        "Compound Score": [compound_score]
    })

demo = gr.Interface(
    fn=analyze_sentiment,
    inputs=["text", "text"],
    outputs="dataframe",
    title="Sentiment Analysis Demo",
    description="Enter text or a URL to analyze sentiment."
)

demo.launch()
```

```python
import os
#Defines a function called extract_article_text that takes a URL as input and
# returns the extracted title and article text.
def extract_article_text(url):
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract article title and text
        title = soup.title.text.strip() if soup.title else "No Title"

        # Identify the main content container (adjust as needed)
        content_container = soup.find('article') or soup.find('div', class_='main-content') or soup.body

        # Extract text only from relevant elements (e.g., paragraphs)
        article_text = ' '.join([p.text for p in content_container.find_all(['p', 'div'], recursive=False)])

        return title, article_text
    except Exception as e:
        print(f"Error extracting content from {url}: {str(e)}")
        return None, None


```

```python
import os
from nltk.corpus import stopwords

def remove_stopwords_from_text(text, stop_words):
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)

def process_files_in_folder(folder_path, stop_words):
    for filename in os.listdir(folder_path):
        filepath = os.path.join(folder_path, filename)

        if os.path.isfile(filepath):
            with open(filepath, 'r', encoding='utf-8') as file:
                text = file.read()

            filtered_text = remove_stopwords_from_text(text, stop_words)

            # Save filtered text back to the same file
            with open(filepath, 'w', encoding='utf-8') as file:
                file.write(filtered_text)

def save_stopwords_to_file(stop_words, output_file_path):
    # Create output folder if it doesn't exist
    os.makedirs(os.path.dirname(output_file_path), exist_ok=True)

    with open(output_file_path, 'w', encoding='utf-8') as output_file:
        for word in stop_words:
            output_file.write(word + '\n')

# Specify the folder path to save the stopwords file
stopwords_output_folder = '/content/drive/MyDrive/sentimental_analysis/stop_words'
stopwords_output_file_path = os.path.join(stopwords_output_folder, 'stopwords.txt')

# Get the stop words list from NLTK
stop_words = set(stopwords.words('english'))

# Save stopwords to a text file
save_stopwords_to_file(stop_words, stopwords_output_file_path)

print("Stopwords saved to:", stopwords_output_file_path)

# Specify the folder path containing text files
text_files_folder_path = '/content/drive/MyDrive/sentimental_analysis'

# Process text files in the folder
process_files_in_folder(text_files_folder_path, stop_words)

print("Stopwords removed and text files saved back to the same folder.")
```

```python
import os
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

def read_stopwords_from_folder(folder_path):
    stop_words = set()

    for filename in os.listdir(folder_path):
        filepath = os.path.join(folder_path, filename)

        if os.path.isfile(filepath):
            with open(filepath, 'r', encoding='utf-8') as file:
                stop_words.update(file.read().split())

    return stop_words

def remove_stopwords_and_create_dictionary(file_path, stop_words):
    word_dictionary = {}

    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()

        # Tokenize the content into words
        words = word_tokenize(content)

        # Remove stop words and create a dictionary
        for word in words:
            word_lower = word.lower()
            if word_lower not in stop_words and word_lower.isalpha():
                word_dictionary[word_lower] = word_dictionary.get(word_lower, 0) + 1

    return word_dictionary

# Specify the folder path containing text files
text_files_folder_path = '/content/drive/MyDrive/sentimental_analysis'

# Specify the folder path containing stop words files
stopwords_folder_path = '/content/drive/MyDrive/sentimental_analysis/stop_words'

# Read stop words from the folder
stop_words = read_stopwords_from_folder(stopwords_folder_path)

# Create a master dictionary by processing each text file
master_dictionary = {}
for filename in os.listdir(text_files_folder_path):
    file_path = os.path.join(text_files_folder_path, filename)

    if os.path.isfile(file_path) and file_path.endswith('.txt'):
        file_dictionary = remove_stopwords_and_create_dictionary(file_path, stop_words)
        master_dictionary.update(file_dictionary)

# Print the results or do something with the master dictionary
print("Master Dictionary:", master_dictionary)
```

```python
import os
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
import string

# Download NLTK resources (if not already downloaded)
nltk.download('punkt')
nltk.download('stopwords')

# Function to count the number of syllables in a word
def count_syllables(word):
    vowels = 'aeiouy'
    count = 0
    prev_char_was_vowel = False
    for char in word.lower():
        if char in vowels and not prev_char_was_vowel:
            count += 1
            prev_char_was_vowel = True
        else:
            prev_char_was_vowel = False
    if word.lower().endswith(('es', 'ed')) and count > 1:
        count -= 1
    return max(1, count)

# Function to determine if a word is complex (having more than 2 syllables)
def is_complex(word):
    return count_syllables(word) > 2

# Function to calculate the average number of words per sentence
def average_words_per_sentence(words, total_sentences):
    return len(words) / total_sentences

# Function to calculate the complex word count
def complex_word_count(words):
    return sum(1 for word in words if is_complex(word))

# Function to clean and count the words in the text
def word_count(text):
    # Tokenize text into words
    words = word_tokenize(text)

    # Remove punctuation and stop words
    stop_words = set(stopwords.words('english'))
    cleaned_words = [word.lower() for word in words if word.lower() not in stop_words and word not in string.punctuation]

    return len(cleaned_words)

# Function to calculate the syllable count per word
def syllable_count_per_word(words):
    syllable_counts = [count_syllables(word) for word in words]
    return syllable_counts

# Function to calculate the count of personal pronouns
def personal_pronoun_count(words):
    personal_pronouns = ['i', 'we', 'my', 'ours', 'us']
    return sum(1 for word in words if word.lower() in personal_pronouns)

# Function to calculate the average word length
def average_word_length(words):
    total_chars = sum(len(word) for word in words)
    return total_chars / len(words)

# Path to the folder containing text files
text_files_folder_path = '/content/drive/MyDrive/sentimental_analysis'

# Initialize counters
total_words = 0
total_sentences = 0
all_words = []

# Iterate through text files in the folder
for file_name in os.listdir(text_files_folder_path):
    if file_name.endswith('.txt'):
        file_path = os.path.join(text_files_folder_path, file_name)

        # Read text from file
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read()

        # Tokenize text into words and sentences
        words = word_tokenize(text)
        sentences = sent_tokenize(text)

        # Update counters
        total_words += len(words)
        total_sentences += len(sentences)
        all_words.extend(words)

# Calculate derived metrics
avg_words_per_sentence = average_words_per_sentence(all_words, total_sentences)
complex_word_cnt = complex_word_count(all_words)
word_cnt = word_count(text)
syllable_cnt_per_word = syllable_count_per_word(all_words)
personal_pronoun_cnt = personal_pronoun_count(all_words)
avg_word_len = average_word_length(all_words)

# Print results
print("Average Number of Words Per Sentence:", avg_words_per_sentence)
print("Complex Word Count:", complex_word_cnt)
print("Word Count:", word_cnt)
print("Syllable Count Per Word:", syllable_cnt_per_word)
print("Personal Pronoun Count:", personal_pronoun_cnt)
print("Average Word Length:", avg_word_len)
```

##1.3Extracting Derived variables
We convert the text into a list of tokens using the nltk tokenize module and use these tokens to calculate the 4 variables described below:

Positive Score: This score is calculated by assigning the value of +1 for each word if found in the Positive Dictionary and then adding up all the values.

Negative Score: This score is calculated by assigning the value of -1 for each word if found in the Negative Dictionary and then adding up all the values. We multiply the score with -1 so that the score is a positive number.

Polarity Score: This is the score that determines if a given text is positive or negative in nature. It is calculated by using the formula:

Polarity Score = (Positive Score – Negative Score)/ ((Positive Score + Negative Score) + 0.000001)
Range is from -1 to +1

Subjectivity Score: This is the score that determines if a given text is objective or subjective. It is calculated by using the formula:

Subjectivity Score = (Positive Score + Negative Score)/ ((Total Words after cleaning) + 0.000001)
Range is from 0 to +1

```python
import os
import json

# Load positive and negative dictionaries from JSON files
with open('/content/drive/MyDrive/sentimental_analysis/positive_dictionary.json', 'r') as f:
    positive_dict = json.load(f)

with open('/content/drive/MyDrive/sentimental_analysis/negative_dictionary.json', 'r') as f:
    negative_dict = json.load(f)

# Function to calculate sentiment scores for a given text
def calculate_sentiment(text):
    # Tokenize the text
    words = text.lower().split()

    # Initialize positive and negative scores
    positive_score = 0
    negative_score = 0

    # Calculate positive and negative scores
    for word in words:
        if word in positive_dict:
            positive_score += 1
        elif word in negative_dict:
            negative_score += 1

    # Calculate polarity score
    polarity_score = (positive_score - negative_score) / ((positive_score + negative_score) + 0.000001)

    # Calculate subjectivity score
    total_words = len(words)
    subjectivity_score = (positive_score + negative_score) / (total_words + 0.000001)

    return positive_score, negative_score, polarity_score, subjectivity_score

# Folder containing text files for sentimental analysis
folder_path = '/content/drive/MyDrive/sentimental_analysis'

# Initialize overall sentiment scores
total_positive_score = 0
total_negative_score = 0

# Iterate through each text file in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):
        with open(os.path.join(folder_path, filename), 'r') as file:
            text = file.read()
            positive_score, negative_score, _, _ = calculate_sentiment(text)
            total_positive_score += positive_score
            total_negative_score += negative_score

# Calculate overall sentiment scores
total_words_count = total_positive_score + total_negative_score
overall_subjectivity_score = total_words_count / (total_words_count + 0.000001)
overall_polarity_score = (total_positive_score - total_negative_score) / ((total_positive_score + total_negative_score) + 0.000001)

print("Total Positive Score:", total_positive_score)
print("Total Negative Score:", total_negative_score)
print("Overall Polarity Score:", overall_polarity_score)
print("Overall Subjectivity Score:", overall_subjectivity_score)
```
